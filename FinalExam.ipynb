```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn

df = pd.read_csv(r"question6.csv", sep=",")
df_US = df[df['Country/Region']=='US']
df_US = df_US[df_US['Province/State']!='US']
df_US['Active'] = df_US['Confirmed'] - df_US['Deaths'] - df_US['Recovered']

df_limited=df_US[["Confirmed","Recovered","Deaths","Active"]]
df_index=df_US["Province/State"]

graph = pd.DataFrame(
    df_limited.to_numpy(), 
    columns=["Confirmed","Recovered","Deaths","Active"],
    index=df_index.to_numpy())
graph.plot.bar(figsize=(20,50));
```
